```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import sem

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug_df = pd.read_csv(mouse_drug_data_to_load)
clinical_trial_df = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
combined_data_df = pd.merge(mouse_drug_df, clinical_trial_df, on="Mouse ID")


# Display the data table for preview

combined_data_df.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
mean_tumor_by_Drug_and_Timepoint_df = combined_data_df.groupby(["Drug","Timepoint"]).mean()["Tumor Volume (mm3)"]
mean_tumor_by_Drug_and_Timepoint_df

# del mean_tumor_by_Drug_and_Timepoint_df["Metastatic Sites"]

# Convert to DataFrame
mean_tumor_by_Drug_and_Timepoint_df_convert = pd.DataFrame(mean_tumor_by_Drug_and_Timepoint_df)


# Preview DataFrame
mean_tumor_by_Drug_and_Timepoint_df_convert.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
standard_error_by_drug_and_timepoint_df = combined_data_df.groupby(["Drug", "Timepoint"]).sem()["Tumor Volume (mm3)"]
standard_error_by_drug_and_timepoint_df

# del standard_error_by_drug_and_timepoint_df["Metastatic Sites"]
# del standard_error_by_drug_and_timepoint_df["Mouse ID"]

# Convert to DataFrame
standard_error_by_drug_and_timepoint_df_convert = pd.DataFrame(standard_error_by_drug_and_timepoint_df)

# Preview DataFrame
standard_error_by_drug_and_timepoint_df_convert.head()

```

```python
# Minor Data Munging to Re-Format the Data Frames
reformat_mean_df = mean_tumor_by_Drug_and_Timepoint_df.reset_index()
reformat_mean_df_pivot = reformat_mean_df.pivot(index="Timepoint", columns="Drug")["Tumor Volume (mm3)"]

reformat_error_df = standard_error_by_drug_and_timepoint_df.reset_index()
reformat_error_df_pivot = reformat_error_df.pivot(index="Timepoint", columns="Drug")["Tumor Volume (mm3)"]

# Preview that Reformatting worked
reformat_mean_df_pivot.head()
reformat_error_df_pivot.head()
```

```python
# Generate the Plot (with Error Bars)
fig, ax = plt.subplots()

# x_axis = 

ax.errorbar(reformat_mean_df_pivot.index, reformat_mean_df_pivot["Capomulin"], yerr=reformat_error_df_pivot["Capomulin"], color="r", marker="o", label="Capomulin")
ax.errorbar(reformat_mean_df_pivot.index, reformat_mean_df_pivot["Infubinol"], yerr=reformat_error_df_pivot["Infubinol"], color="b", marker="^", label="Infubinol")
ax.errorbar(reformat_mean_df_pivot.index, reformat_mean_df_pivot["Ketapril"], yerr=reformat_error_df_pivot["Ketapril"], color="g", marker="s", label="Ketapril")
ax.errorbar(reformat_mean_df_pivot.index, reformat_mean_df_pivot["Placebo"], yerr=reformat_error_df_pivot["Placebo"], color="k", marker="d", label="Placebo")

plt.title("Tumor Response to Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume (mm3)")

plt.grid()
plt.legend()

# Save the Figure
plt.savefig("Tumor_Response_to_Treatment.png")

# Show the Figure
plt.show()

```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
mean_metsite_by_Drug_and_Timepoint_df = combined_data_df.groupby(["Drug","Timepoint"]).mean()["Metastatic Sites"]
mean_metsite_by_Drug_and_Timepoint_df.head()

# Convert to DataFrame
mean_metsite_by_Drug_and_Timepoint_df_convert = pd.DataFrame(mean_metsite_by_Drug_and_Timepoint_df)


# Preview DataFrame
mean_metsite_by_Drug_and_Timepoint_df_convert.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
standard_error_metsite_by_Drug_and_Timepoint_df = combined_data_df.groupby(["Drug","Timepoint"]).sem()["Metastatic Sites"]
standard_error_metsite_by_Drug_and_Timepoint_df.head()

# Convert to DataFrame
standard_error_metsite_by_Drug_and_Timepoint_df_convert = pd.DataFrame(standard_error_metsite_by_Drug_and_Timepoint_df)
# Preview DataFrame
standard_error_metsite_by_Drug_and_Timepoint_df_convert.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
reformat_mean_df_metastatic = mean_metsite_by_Drug_and_Timepoint_df_convert.reset_index()
reformat_mean_df_metastatic_pivot = reformat_mean_df_metastatic.pivot(index="Timepoint", columns="Drug")["Metastatic Sites"]

reformat_error_df_metastatic = standard_error_metsite_by_Drug_and_Timepoint_df_convert.reset_index()
reformat_error_df_metastatic_pivot = reformat_error_df_metastatic.pivot(index="Timepoint", columns="Drug")["Metastatic Sites"]


# Preview that Reformatting worked
reformat_mean_df_metastatic_pivot.head()
reformat_error_df_metastatic_pivot.head()
```

```python
# Generate the Plot (with Error Bars)
fig, ax = plt.subplots()


ax.errorbar(reformat_mean_df_metastatic_pivot.index, reformat_mean_df_metastatic_pivot["Capomulin"], yerr=reformat_error_df_metastatic_pivot["Capomulin"], color="r", marker="o", label="Capomulin")
ax.errorbar(reformat_mean_df_metastatic_pivot.index, reformat_mean_df_metastatic_pivot["Infubinol"], yerr=reformat_error_df_metastatic_pivot["Infubinol"], color="b", marker="^", label="Infubinol")
ax.errorbar(reformat_mean_df_metastatic_pivot.index, reformat_mean_df_metastatic_pivot["Ketapril"], yerr=reformat_error_df_metastatic_pivot["Ketapril"], color="g", marker="s", label="Ketapril")
ax.errorbar(reformat_mean_df_metastatic_pivot.index, reformat_mean_df_metastatic_pivot["Placebo"], yerr=reformat_error_df_metastatic_pivot["Placebo"], color="k", marker="d", label="Placebo")

plt.title("Metastatic Spraed During Treatment")
plt.xlabel("Treatment Duration (Days)")
plt.ylabel("Metastic Sites")

plt.grid()
plt.legend()

# Save the Figure
plt.savefig("Metastatic_Spread_During_Treatment.png")

# Show the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
count_mice_by_drug_and_timepoint = combined_data_df.groupby(["Drug", "Timepoint"]).count()["Mouse ID"]
# count_mice_by_drug_and_timepoint.head()

# Convert to DataFrame
count_mice_by_drug_and_timepoint_convert = pd.DataFrame(count_mice_by_drug_and_timepoint)
# Preview DataFrame
count_mice_by_drug_and_timepoint_convert.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
reformat_count_mice_by_drug_and_timepoint = count_mice_by_drug_and_timepoint_convert.reset_index()
reformat_count_mice_by_drug_and_timepoint_pivot = reformat_count_mice_by_drug_and_timepoint.pivot(index="Timepoint", columns="Drug")["Mouse ID"]
# Preview the Data Frame
reformat_count_mice_by_drug_and_timepoint_pivot.head()
```

```python
# Generate the Plot (Accounting for percentages)
percentage_capomulin = 100 * reformat_count_mice_by_drug_and_timepoint_pivot["Capomulin"] / 25
percentage_infubinol = 100 * reformat_count_mice_by_drug_and_timepoint_pivot["Infubinol"] / 25
percentage_ketapril = 100 * reformat_count_mice_by_drug_and_timepoint_pivot["Ketapril"] / 25
percentage_placebo = 100 * reformat_count_mice_by_drug_and_timepoint_pivot["Placebo"] / 25

plt.plot(percentage_capomulin, color="r", marker="o", label="Capomulin")
plt.plot(percentage_infubinol, color="b", marker="^", label="Infubinol")
plt.plot(percentage_ketapril, color="g", marker="s", label="Ketapril")
plt.plot(percentage_placebo, color="k", marker="d", label="Placebo")

plt.title("Survival During Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Survival Rate (Percentage)")

plt.grid()
plt.legend()

# Save the Figure
plt.savefig("Survival_During_Treatment.png")

# Show the Figure
plt.show()
```
